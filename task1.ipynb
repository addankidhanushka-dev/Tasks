```python
# q7    Find the team which won the match by the highest number of runs and the lowest number of runs.

matches = pd.read_csv("matches.csv")

# consider only matches won by runs
runs_wins = matches[matches["win_by_runs"] > 0]

# highest and lowest run margins
max_runs = runs_wins["win_by_runs"].max()
min_runs = runs_wins["win_by_runs"].min()

# all teams with highest run wins
highest_run_wins = runs_wins[runs_wins["win_by_runs"] == max_runs]

# all teams with lowest run wins
lowest_run_wins = runs_wins[runs_wins["win_by_runs"] == min_runs]

print("Teams with highest win by runs:")
highest_run_wins[["winner", "win_by_runs"]]

```

```python
# q7 -- continued

print("\nTeams with lowest win by runs:")
lowest_run_wins[["winner", "win_by_runs"]]
```

```python
# q8    Calculate the mean, median, and standard deviation of win_by_runs.
df = pd.read_csv("matches.csv")

mean_val = df["win_by_runs"].mean()
median_val = df["win_by_runs"].median()
std_val = df["win_by_runs"].std()

print(f"Mean of win_by_runs: {round(mean_val, 2)}")
print(f"Median of win_by_runs: {round(median_val, 2)}")
print(f"Standard Deviation of win_by_runs: {round(std_val, 2)}")


```

```python
# q12   Compute the average runs scored in matches at all venues.
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

# runs scored in each match
match_runs = (
    deliveries
    .groupby('match_id')['total_runs']
    .sum()
    .reset_index()
)

# Merge venue information
match_runs = match_runs.merge(
    matches[['id', 'venue']],
    left_on='match_id',
    right_on='id'
)

# Average runs per match at each venue
avg_runs_per_venue = (
    match_runs
    .groupby('venue')['total_runs']
    .mean()
    .reset_index()
)

# Display result
print(avg_runs_per_venue)
```

```python
# q13   Find the umpires who umpired the maximum number of matches.

# Load dataset
matches = pd.read_csv('matches.csv')

# Combine both umpire columns into one Series
all_umpires = pd.concat([                           # pd.concat() stacks them vertically (row-wise).
    matches['umpire1'],
    matches['umpire2'],
    matches['umpire3']
])

# Count how many times each umpire officiated
umpire_counts = all_umpires.value_counts()

# Find maximum number of matches umpired
max_matches = umpire_counts.max()

# Umpires(series) who umpired maximum times
top_umpires = umpire_counts[umpire_counts == max_matches]

print(top_umpires)
```

```python

# q14   Find the total number of matches played in each season.
matches = pd.read_csv('matches.csv')


matches_per_season = (
    matches
    .groupby('season')
    .size()                                         # .size() counts rows whereas .count() counts non-null values
    .reset_index(name='total_matches')
)

print(matches_per_season)
```

```python
# q15   Find the total runs scored in each season.
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

# Merge deliveries with season info                     For each delivery, find the match with the same match id
                                                        # and attach its season
data = deliveries.merge(                                # Number of rows = same as deliveries
    matches[['id', 'season']],                          # deliveries.match_id == matches.id ,
    left_on='match_id',                                 # col used for deliveries df
    right_on='id'                                       # col used for matches df
)

# Total runs scored in each season
runs_per_season = (
    data
    .groupby('season')['total_runs']
    .sum()
    .reset_index(name='total_runs')
)

# Display result
print(runs_per_season)
```

```python
# q16   Calculate the total number of runs scored by each batsman and display the top 10 batsmen.

deliveries = pd.read_csv("deliveries.csv")

group = deliveries.groupby("batsman")

batsman_runs = group["batsman_runs"].sum()

top_10_batsmen = batsman_runs.sort_values(ascending=False).head(10)

print(top_10_batsmen)
```

```python
# q17   Compute the total number of wickets taken by each bowler.
deliveries = pd.read_csv("deliveries.csv")

# filter only valid wicket events (exclude run out, retired hurt, etc.)
wickets = deliveries[
    deliveries["dismissal_kind"].notna() &                          # keeps only deliveries where a wicket fell
    (deliveries["dismissal_kind"] != "run out")
]

bowler_wickets = wickets.groupby("bowler").size()

# descending order
bowler_wickets = bowler_wickets.sort_values(ascending=False)

print(bowler_wickets)
```

```python
# q18   Compute batting averages and display the top 10 batsmen.

# Batting average = total runs scored (runs)/ number of times dismissed (outs)
deliveries = pd.read_csv("deliveries.csv")

# total runs per batsman
runs = deliveries.groupby("batsman")["batsman_runs"].sum()

# number of times each batsman got out
outs = deliveries[deliveries["player_dismissed"].notna()] \
        .groupby("player_dismissed").size()

# combine runs and outs
batting_avg = pd.concat([runs, outs], axis=1)
batting_avg.columns = ["runs", "outs"]

# remove batsmen with zero dismissals
batting_avg = batting_avg[batting_avg["outs"] > 0]

# calculate batting average
batting_avg["average"] = batting_avg["runs"] / batting_avg["outs"]

# top 10 batting averages
top_10 = batting_avg.sort_values("average", ascending=False).head(10)

print(top_10)
```

```python
# q19   Visualize toss decisions across all seasons.
import matplotlib.pyplot as plt

matches = pd.read_csv("matches.csv")

# season vs toss decision table
season_toss = pd.crosstab(matches["season"], matches["toss_decision"])

# plot side-by-side bars
season_toss.plot(kind="bar", figsize=(10,5))

plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.title("Season-wise Toss Decisions (Bat vs Field)")
plt.legend(title="Toss Decision")
plt.tight_layout()
plt.show()
```

```python
# q21   Find the distribution of teams that won the matches.


matches = pd.read_csv("matches.csv")

# count wins per team
win_distribution = matches["winner"].dropna().value_counts()

# create wrapped labels (two lines)
labels = [team.replace(" ", "\n") for team in win_distribution.index]

# plot
plt.figure(figsize=(12, 6))
plt.bar(labels, win_distribution.values)

plt.xlabel("Teams")
plt.ylabel("Number of Matches Won")
plt.title("Distribution of Matches Won by Teams")

plt.tight_layout()
plt.show()
```

```python
# q22   Visualize the toss outcomes of all teams.

matches = pd.read_csv("matches.csv")

# count toss decisions for each team
toss_outcomes = (
    matches
    .groupby(["toss_winner", "toss_decision"])
    .size()
    .unstack(fill_value=0)
)

# plotting setup
x = np.arange(len(toss_outcomes.index))
width = 0.4

fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(x - width/2, toss_outcomes["bat"], width, label="Bat")
ax.bar(x + width/2, toss_outcomes["field"], width, label="Field")

ax.set_xlabel("Teams")
ax.set_ylabel("Number of Tosses")
ax.set_title("Toss Outcomes of All Teams")

ax.set_xticks(x)
ax.set_xticklabels(
    [team.replace(" ", "\n") for team in toss_outcomes.index],
    rotation=0
)

ax.legend(loc="upper right")

plt.tight_layout()
plt.show()


```
